# Pandas 자료형

```python
import pandas as pd
import seaborn as sns #그래프
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
#열 이름 변경
tips = tips.rename({'sex' : 'gender'}, axis=1) #axis='columns'
tips
```

```python
#열의 데이터 타입 변경 : astype()
print(tips.dtypes); print()
tips['total_bill'] = tips['total_bill'].astype(str)
print(tips.dtypes); print()
tips['total_bill'] = tips['total_bill'].astype('float')
print(tips.dtypes); print()
```

```python
#데이터 변경
import warnings
warnings.filterwarnings(action='ignore')

tips_sub_miss = tips.head(10)
tips_sub_miss.loc[1:8:2,'total_bill'] = 'missing'
tips_sub_miss
```

```python
print(tips_sub_miss.dtypes); print()
# tips_sub_miss['total_bill'] = tips_sub_miss['total_bill'].astype(float)
# print(tips_sub_miss.dtypes); print()
```

```python
#열의 데이터 타입 변경(to_numeric)
#옵션 raise : 숫자로 변환할 수 없는 값 있으면 오류 발생
#     coerce : 숫자로 변환할 수 없는 값 있으면 누락값 대입(NaN)
#     ignore : 숫자로 변환할 수 없는 값 있으면 오류 발생
tips_sub_miss['total_bill'] = pd.to_numeric(tips_sub_miss['total_bill'], errors='coerce')
print(tips_sub_miss.dtypes); print()

tips_sub_miss
```

```python
#다운캐스팅 변환
tips_sub_miss['total_bill'] = pd.to_numeric(tips_sub_miss['total_bill'], errors='coerce', downcast='float')
print(tips_sub_miss.dtypes)
```
